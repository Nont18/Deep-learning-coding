```python
import os
import cv2
import mediapipe as mp
import numpy as np

# Initialize MediaPipe Holistic
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic(static_image_mode=False, min_detection_confidence=0.5)

```

```python
for i in range(len(video_list)):
    print("Processing video: {}. {} out of {} videos processed".format(video_list[i], i+1, len(video_list)))
    
```

```python
# Function to process video frames and extract keypoints
def extract_keypoints_from_video(video_path, num_frames=30):
    cap = cv2.VideoCapture(video_path)
    keypoints_all_frames = []
    frame_count = 0

    while cap.isOpened() and frame_count < num_frames:
        ret, frame = cap.read()
        if not ret:
            break

        # Convert BGR to RGB
        image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = holistic.process(image_rgb)

        # Extract keypoints from pose, face, and hands
        pose_landmarks = results.pose_landmarks.landmark if results.pose_landmarks else []
        face_landmarks = results.face_landmarks.landmark if results.face_landmarks else []
        left_hand_landmarks = results.left_hand_landmarks.landmark if results.left_hand_landmarks else []
        right_hand_landmarks = results.right_hand_landmarks.landmark if results.right_hand_landmarks else []

        # Flatten the landmark data
        keypoints = np.array(
            [item for sublist in [
                pose_landmarks, face_landmarks, left_hand_landmarks, right_hand_landmarks
            ] for item in sublist]
        ).flatten()

        # Ensure the array has 1662 values (543 keypoints * 3 coordinates)
        if len(keypoints) == 1662:
            keypoints_all_frames.append(keypoints)
            frame_count += 1

    cap.release()

    # If fewer than 30 frames were captured, pad with zero arrays
    while len(keypoints_all_frames) < num_frames:
        keypoints_all_frames.append(np.zeros(1662))

    return np.array(keypoints_all_frames)
```
