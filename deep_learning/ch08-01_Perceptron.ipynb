### Artificial Neuron / Perceptron
สร้าง AND Gate ด้วย Neuron (Perceptron) ด้วย Scikit-learn

```python
import numpy as np 

X = [[0, 0],
    [0, 1],
    [1, 0],
    [1, 1]]

X_train = np.array(X)

y = [0, 0, 0, 1] # target (y) for AND gate
#y = [0,1,1,1] # OR gate
#y = [0,1,1,0] # XOR
y_train = np.array(y)
```

### Train

```python
from sklearn.linear_model import Perceptron

# model = Perceptron(random_state=0, verbose=False)
model = Perceptron(random_state=0, verbose=True)
# model = Perceptron()

model.fit(X_train, y_train)
```

### Evaluate

```python
print('score: ', model.score(X_train, y_train))
print('predicted: ', model.predict(X_train))
print('target: ', np.array(y))
```

### Weight and bias

```python
print('weight [w1 w2] =', model.coef_[0])
print('w0 (bias) =', model.intercept_[0])
```

### Perceptron ด้วย Python ไม่ใช้ Library (Scratch Code)

```python
def and_gate(x1, x2):
    w0 = -2 #bias
    w1 = 2
    w2 = 2

    threshold = 0.0

    s = w0 + (w1 * x1) + (w2 * x2)

    if s> threshold:
        result = 1
    else:
        result = 0

    print('output y={} [s={:.2f}]'.format(result, s))

and_gate(0,0)
and_gate(0,1)
and_gate(1,0)
and_gate(1,1)
```
