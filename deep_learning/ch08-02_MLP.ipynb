### MLP (XOR)

```python
import numpy as np 

X = [[0,0],
    [0,1],
    [1,0],
    [1,1]]

X_train = np.array(X)

y = [0,1,1,0]
y_train = np.array(y)
```

### Train Model

```python
from sklearn.neural_network import MLPClassifier

# model = MLPClassifier(hidden_layer_sizes=(3), random_state=1, verbose=True)
# model = MLPClassifier(hidden_layer_sizes=(3))

# model = MLPClassifier(hidden_layer_sizes=(4), random_state=1, verbose=True)
model = MLPClassifier(hidden_layer_sizes=(300), random_state=1, verbose=True, max_iter=900)

model.fit(X_train, y_train)
```

### Evaluation

```python
print('score : ', model.score(X_train, y_train))
print('predicted : ', model.predict(X_train))
print('target : ', np.array(y_train))
```

### Model weight

```python
[coef.shape for coef in model.coefs_]
```

```python
print(model.coefs_) #ค่า weight
```

```python
print(model.intercepts_) # ค่า bias
```

### Plot Decision Regions

```python
import matplotlib.pyplot as plt 
import mglearn

plt.figure(figsize=(5,4))
mk = {0: 'o', 1: 's'} # or mk = ['o','s'] #work

mglearn.plots.plot_2d_separator(model, X_train, fill=True, eps=1.5, alpha=0.28)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y, s=12, markers=mk)

#plt.legend(loc='upper left')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper left')
plt.title('MLP Perceptron')
plt.show()
```
