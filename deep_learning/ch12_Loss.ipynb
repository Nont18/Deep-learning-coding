### Binary Cross-Entropy

```python
import numpy as np 
import tensorflow as tf 

y_true = [0, 0, 1, 1]
y_pred = [0, 0, 1, 1]
```

```python
y_true = [0, 0, 1, 1]
y_pred = [0, 0, 1, 0]
```

```python
y_true = [0, 0, 1, 1]
y_pred = [1, 1, 0, 0]
```

```python
y_pred = np.array(y_pred).astype('float')
```

```python
bce = tf.keras.losses.BinaryCrossentropy()
bce(y_true, y_pred).numpy()
```

```python
import matplotlib.pyplot as plt 

def cross_entropy_loss(yHat, y):
    if y == 1:
        return -np.log(yHat)
    else:
        return -np.log(1 - yHat)
    
y_pred = np.arange(0.01, 0.999, 0.01)

cost_1 = cross_entropy_loss(y_pred, 1)
cost_0 = cross_entropy_loss(y_pred, 0)

plt.plot(y_pred, cost_1, label='Actual (y_true) = 1', c='b', ls='-', lw=2)
plt.plot(y_pred, cost_0, label='Actual (y_true) = จ', c='r', ls='--', lw=2)

plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
