### AND Gate using MultiLayer Perceptron

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Dataset

```python
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,0,0,1]

X = np.array(X)
y = np.array(y)
```

```python
print('X =', X)
print('y =', y)
```

### Data Visulalization

### Model and Train

```python
# from sklearn.linear_model import LinearRegression
# from sklearn.tree import DecisionTreeClassifier
# from sklearn.linear_model import Perceptron
# from sklearn.neural_network import MLPClassifier
# from sklearn.preprocessing import StandardScaler #ใช้กัย data ที่มีการแจกแจงแบบ normal
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
iter = 200
hidden = 100
model = MLPClassifier(random_state=1, verbose=True, hidden_layer_sizes=hidden, max_iter=iter)
```

```python
model.fit(X, y)
```

```python
model.predict(X)
```

### Evaluation

```python
#แสดงผล
print("score : ", model.score(X,y))
print("Predict : ", model.predict(X))
print("Target : ", np.array(y))
```

### Plot Decision Regions

```python
import matplotlib.pyplot as plt 
import mglearn

plt.figure(figsize=(5,4))
mk = ['o', 's']

mglearn.plots.plot_2d_separator(model, X, fill=True, eps=1.5, alpha=0.28)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y, s=12, markers=mk)

#plt.legend(loc='upper left')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper left')
plt.title('MLP Perceptron')
plt.show()
```
