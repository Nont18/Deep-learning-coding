### Mushroom Classification

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Dataset

```python
df = pd.read_csv('data/mushroom_cleaned.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.season.unique()
```

```python
X = df.drop(['class', 'stem-height', 'stem-width', 'cap-diameter', 'season'], axis=1)
X
```

```python
y = df.drop(['cap-diameter', 'cap-shape', 'gill-attachment', 'gill-color', 'stem-height', 'stem-width', 'stem-color', 'season'], axis=1)
y
```

```python
X = np.array(X)
X
```

```python
y = np.array(y)
y
```

### Data Visulization

```python
# plt.scatter(df.season, df.class)
```

### Model and Train

```python
from sklearn.neural_network import MLPClassifier
```

```python
iter = 100
hidden = 300
model = MLPClassifier(random_state=1, verbose=True, hidden_layer_sizes=hidden, max_iter=iter)
```

```python
model.fit(X,y)
```

### Predict

```python
model.predict(X)
```

```python
df.sample()
```

```python
# Feature = [cap-shape, gill-attachment, gill-color, stem-color]
X_input = [ 6,	4,	3, 11]
X_input = np.array(X_input)
X_input = X_input.reshape(-1,4)
X_input
```

```python
y_predict = model.predict(X_input)
y_predict
```

### Evaluation

```python
model.score(X,y)
```

```python
y_predict = model.predict(X)
```

```python
class_names = ['Poisonous', 'Eatable']
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y, y_predict)
ConfusionMatrixDisplay(cm, display_labels=class_names).plot()
plt.title('Confusion Matrix')
plt.show()
```

### Decision Region

```python
import matplotlib.pyplot as plt 
import mglearn

plt.figure(figsize=(5,4))
mk = {0: 'o', 1: 's'} # or mk = ['o','s'] #work

mglearn.plots.plot_2d_separator(model, X, fill=True, eps=1.5, alpha=0.28)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y, s=12, markers=mk)

#plt.legend(loc='upper left')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper left')
plt.title('MLP Perceptron')
plt.show()
```

### Loss Curve

```python
plt.figure(figsize=(5,3))
plt.title('Loss curve')
plt.xlabel('Iteration')
plt.plot(model.loss_curve_)
plt.tight_layout()
plt.show()
```
