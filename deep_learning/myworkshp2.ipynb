### Mobile Price Prediction

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/Mobile-Price-Prediction-cleaned_data.csv')
df.head()
```

```python
df.shape
```

```python
X1 = df.Ratings
X1 = np.array(X1)
X1
```

```python
X2 = df.RAM
X2 = np.array(X2)
X2
```

```python
X3 = df.Battery_Power
X3 = np.array(X3)
X3
```

```python
X4 = df.Mobile_Size
X4 = np.array(X4)
X4
```

```python
X5 = df.ROM
X5 = np.array(X5)
X5
```

```python
data = list(zip(X1,X2,X3,X4,X5))
X = pd.DataFrame(data)
X
```

```python
cols = ['Ratings', 'RAM', 'Battery_Power', 'Mobile_Size', 'ROM']
X = pd.DataFrame(data, columns=cols)
X
```

```python
y = df.Price
y
```

```python
y = np.array(y)
```

```python
X = np.array(X)
X = X.reshape(-1,5)
X
```

### Data Visualization

```python
plt.scatter(df.RAM, df.Price)

plt.xlabel('X')
plt.ylabel('y')
plt.title('Visualization')
plt.show()
```

### Train and Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# ['Ratings', 'RAM', 'Battery_Power', 'Mobile_Size', 'ROM']
X_input = [4.5, 2, 4200, 5.5, 16]
X_input = np.array(X_input)
X_input = X_input.reshape(-1,5)
X_input
```

```python
y_predict = model.predict(X_input)
y_predict
```

```python
print('score: ', model.score(X, y))
print('predicted: ', model.predict(X))
print('target: ', np.array(y))
```
